### Plotting

```
def fractions(x,pos):
    if np.isclose((x/step)%(1./step),0.):        
        if x == 1:
          return '$\pi$'.format(x) 
          
        return '${:.0f} \pi$'.format(x)
    else:        
        return '$\\frac{{{:2.0f}\pi}}{{{:2.0f}}}$'.format(x/step,1./step)        
```

```
import matplotlib.ticker as tck

fig = plt.figure(figsize=(14,10))
gs = fig.add_gridspec(1, 1)
ax1 = fig.add_subplot(gs[0, 0])

# ax1.xaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))
step = 1/4
ax1.xaxis.set_major_locator(tck.MultipleLocator(step))
ax1.xaxis.set_major_formatter(tck.FuncFormatter(fractions))
for tick in ax1.xaxis.get_major_ticks():
    tick.label.set_fontsize(20) 
for tick in ax1.yaxis.get_major_ticks():
    tick.label.set_fontsize(20) 

fig.suptitle('Fidelity', size=24)

ax1.plot(x_pi, AE_fid, 'r', label='AE')
ax1.plot(x_pi, AE_val_fid, 'r:')

ax1.plot(x_pi, TAE_fid, 'b', label='TAE')
ax1.plot(x_pi, TAE_val_fid, 'b:')

ax1.set_xlabel('Error', size=20)
ax1.set_ylabel('Fidelity', size=20)
ax1.legend(prop={'size':20})

plt.show()
```
