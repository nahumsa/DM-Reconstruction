# Analysing AutoEncoders

## Importing the values

```
AE_Fidelity = np.loadtxt('Results/AE/AE_Fidelity.txt')
AE_Loss = np.loadtxt('Results/AE/AE_Loss.txt')
AE_val_Fidelity = np.loadtxt('Results/AE/AE_val_Fidelity.txt')
AE_val_Loss = np.loadtxt('Results/AE/AE_val_Loss.txt')
```

```
TAE_Fidelity = np.loadtxt('Results/TAE/TAE_Fidelity.txt')
TAE_Loss = np.loadtxt('Results/TAE/TAE_Loss.txt')
TAE_val_Fidelity = np.loadtxt('Results/TAE/val_TAE_Fidelity.txt')
TAE_val_Loss = np.loadtxt('Results/TAE/val_TAE_Loss.txt')
```

```
TVAE_Fidelity = np.loadtxt('Results/TVAE/TVAE_Fidelity.txt')
TVAE_Loss = np.loadtxt('Results/TVAE/TVAE_Loss.txt')
TVAE_val_Fidelity = np.loadtxt('Results/TVAE/val_TVAE_Fidelity.txt')
TVAE_val_Loss = np.loadtxt('Results/TVAE/val_TVAE_Loss.txt')
```

```
epochs = range(len(TAE_Fidelity))
```

## Plotting

```
fig = plt.figure(figsize=(14,10))
gs = fig.add_gridspec(1, 1)
ax1 = fig.add_subplot(gs[0, 0])


fig.suptitle('Fidelity', size=16)

ax1.plot(epochs, AE_Fidelity, 'r', label='AE')
ax1.plot(epochs, TAE_Fidelity, 'b', label='TAE')
ax1.plot(epochs, TVAE_Fidelity, 'g', label='TVAE')
ax1.plot(epochs, AE_val_Fidelity, 'r:')
ax1.plot(epochs, TAE_val_Fidelity, 'b:')
ax1.plot(epochs, TVAE_val_Fidelity, 'g:')
# ax1.set_title('Fidelity')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Fidelity')
ax1.legend()

plt.show()
```

```
fig = plt.figure(figsize=(14,5))
gs = fig.add_gridspec(1, 2)
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])

fig.suptitle('Fidelity', size=16)

ax1.plot(epochs, AE_Fidelity, 'r', label='AE')
ax1.plot(epochs, TAE_Fidelity, 'b', label='TAE')
ax1.plot(epochs, TVAE_Fidelity, 'g', label='TVAE')
ax1.set_title('Training Data')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Fidelity')
ax1.legend()

ax2.plot(epochs, AE_val_Fidelity, 'r', label='AE')
ax2.plot(epochs, TAE_val_Fidelity, 'b', label='TAE')
ax2.plot(epochs, TVAE_val_Fidelity, 'g', label='TVAE')
ax2.set_title('Test Data')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Fidelity')
ax2.legend()

plt.show()
```

```
fig = plt.figure(figsize=(14,5))
gs = fig.add_gridspec(1, 2)
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])

fig.suptitle('Loss', size=16)
ax1.plot(epochs, AE_Loss, 'r', label='AE')
ax1.plot(epochs, TAE_Loss, 'b', label='TAE')
ax1.plot(epochs, TVAE_Loss/10, 'g', label='TVAE')
ax1.set_title('Training Data')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Fidelity')
ax1.legend()

ax2.plot(epochs, AE_val_Loss, 'r', label='AE')
ax2.plot(epochs, TAE_val_Loss, 'b', label='TAE')
ax2.plot(epochs, TVAE_val_Loss/10, 'g', label='TVAE')
ax2.set_title('Test Data')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Fidelity')
ax2.legend()

plt.show()
```
